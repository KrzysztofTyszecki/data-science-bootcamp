```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot([0, 1], [0, 3], label='awesome_green_line', color='green')
plt.legend()
plt.xlabel('X axe')
plt.ylabel('Y axe')
plt.title('Awesome Line')
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
```

```python
plt.plot(x, x_sin, label='sinus')
plt.plot(x, x_cos, label='cosinus')
plt.legend()
```
